```python
import shutil
import os
import time
import schedule
import git
import subprocess

def backupLocal(path='..\\backup'):
    files=[file for file in os.listdir('.') if os.path.isfile(file)]
    if not os.path.exists(path):
        os.makedirs(path)
    date=time.strftime('%Y%m%d%H%M%S')
    os.makedirs(f'{path}\\{date}')
    for file in files:
        shutil.copy(file,f'{path}\\{date}\\{file}')

'''def backupGitHub(path='C:\\Users\\gsr00\\软工\\code'):
    try:
        repo=git.Repo(path)
    except:
        repo.git.Repo.init(path)
    repo.git.add('-A')
    date=time.strftime('%Y.%m.%d %H:%M:%S')
    commit_message=f'Automatically update at {date}'
    repo.index.commit(commit_message)
    origin=repo.remote(name='origin')
    ssh_key_path = 'C:\\Users\\gsr00\\.ssh\\id_rsa'  # 修改为你的私钥路径
        
        # 方法1: 使用环境变量指定SSH密钥
    with repo.git.custom_environment(GIT_SSH_COMMAND=f'ssh -i {ssh_key_path}'):
        print("开始推送...")
        push_info = origin.push()
        for info in push_info:
            print(f"推送结果: {info.summary}")
        
    print('推送成功')'''

def backupGitHub_subprocess(path='C:\\Users\\gsr00\\软工\\code'):
    try:
        # 切换到仓库目录
        original_dir = os.getcwd()
        os.chdir(path)
        
        # 添加和提交
        subprocess.run(['git', 'add', '-A'], check=True)
        date = time.strftime('%Y.%m.%d %H:%M:%S')
        commit_message = f'Automatically update at {date}'
        subprocess.run(['git', 'commit', '-m', commit_message], check=True)
        
        # 推送并显示实时输出
        print("开始推送...")
        process = subprocess.Popen(
            ['git', 'push', '--progress'],
            stdout=subprocess.PIPE,
            stderr=subprocess.STDOUT,
            universal_newlines=True
        )
        
        # 实时读取输出
        while True:
            output = process.stdout.readline()
            if output == '' and process.poll() is not None:
                break
            if output:
                print(output.strip())
        
        return_code = process.poll()
        if return_code == 0:
            print('推送成功')
        else:
            print(f'推送失败，返回码: {return_code}')
            
        os.chdir(original_dir)
        
    except Exception as e:
        print(f'备份失败: {e}')
        os.chdir(original_dir)

if __name__=='__main__':
    #backupGitHub()
    #backupLocal()
    backupGitHub_subprocess()
```
